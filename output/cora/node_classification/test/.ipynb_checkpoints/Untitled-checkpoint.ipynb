```python
import json

jsonl_path = "2_hop_without_label.jsonl"  # 改成你的 jsonl 文件路径

with open(jsonl_path, "r", encoding="utf-8") as f:
    for line_num, line in enumerate(f, 1):
        line = line.strip()
        if not line:
            continue

        try:
            item = json.loads(line)
        except json.JSONDecodeError:
            continue

        context_value = item.get("Context", "")
        if isinstance(context_value, str) and "1485" in context_value:
            print(f"在第 {line_num} 行找到目标 item：")
            print(json.dumps(item, ensure_ascii=False, indent=2))
            break
    else:
        print("没有找到匹配的 item")
```
