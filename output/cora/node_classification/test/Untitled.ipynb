```python
import json

jsonl_path = "2_hop_without_label.jsonl"  # 改成你的文件路径
target_str = "1485"

matched_lines = []

with open(jsonl_path, "r", encoding="utf-8") as f:
    for line_num, line in enumerate(f, 1):
        line = line.strip()
        if not line:
            continue

        try:
            item = json.loads(line)
        except json.JSONDecodeError:
            continue

        context_value = item.get("Context", "")
        if isinstance(context_value, str) and target_str in context_value:
            matched_lines.append(line_num)

if matched_lines:
    print("所有匹配的行号：")
    print(matched_lines)
else:
    print("没有找到匹配项")
```
