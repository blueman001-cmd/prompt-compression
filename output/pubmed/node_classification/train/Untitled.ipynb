```python
import json
from collections import defaultdict

input_path = "2_hop_without_label.jsonl"      # 你的原始文件
output_path = "2_hop_without_label_10.jsonl"  # 过滤后的文件

max_per_answer = 10
answer_count = defaultdict(int)

kept = 0
skipped = 0

with open(input_path, "r", encoding="utf-8") as fin, \
     open(output_path, "w", encoding="utf-8") as fout:
    
    for line in fin:
        line = line.strip()
        if not line:
            continue

        try:
            item = json.loads(line)
        except json.JSONDecodeError:
            print(f"跳过非法 JSON 行: {line[:100]}")
            skipped += 1
            continue

        if "Answer" not in item:
            print(f"跳过缺少 Answer 的 item: {item}")
            skipped += 1
            continue

        answer = item["Answer"]

        if answer_count[answer] < max_per_answer:
            fout.write(json.dumps(item, ensure_ascii=False) + "\n")
            answer_count[answer] += 1
            kept += 1
        else:
            skipped += 1

print("处理完成")
print(f"保留条数: {kept}")
print(f"跳过条数: {skipped}")
print("各 Answer 保留数量:")
for ans, cnt in answer_count.items():
    print(f"{ans}: {cnt}")
```
